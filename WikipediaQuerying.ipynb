```python
import pandas as pd
import wikipedia as wk

data = pd.read_csv("data/processed_recs2.csv")
df = pd.DataFrame(data=data)
df.head()
```

```python
df = df.loc[~(df["scholar_id"] == "na")]
df
```

```python
len(df.index)
```

```python
f = df[df["gender"] == "female"].copy()
f["citedby"]
```

```python
def cited_to_num(x):
    try:
        x["citedby"] = float(x["citedby"])
    except:
        x["citedby"] = "na"
    return x;

f = f.apply(cited_to_num, axis=1)

#bins = [0, 25, 50, 75, 100]
#bin_labels = ["0_25", "25_50", "50_75", "75_100"]

categories = pd.qcut(f["citedby"], 3, labels=["1_3", "2_3", "3_3"])
f["categories"] = pd.qcut(f["citedby"], 3, labels=["1_3", "2_3", "3_3"])
#categories
retbins = pd.qcut(f["citedby"], 3, labels=["1_3", "2_3", "3_3"], retbins=True)
retbins
```

```python
def get_wikipage(x):
    try:
        page = wk.page(x["c_fname"] + " " + x["c_lname"] + " Nature")
        x["page_id"] = page.pageid
        x["has_page"] = 1
    except:
        x["page_id"] = "na"
        x["has_page"] = 0
        
    return x;

df_wk = df.apply(get_wikipage, axis=1)
df_wk
    
```

```python
df_wk.to_csv("data/wiki_recs2.csv")
```
