```python
import pandas as pd
import requests
import scholarly 
import time
import gscholar

# Downloaded as tab-deliminated file 
#then converted to CSV in Excel
data = pd.read_csv("data/savedrecs4.csv")
#pd.read_tsv("data/savedrecs (2).txt")
df = pd.DataFrame(data=data)

# Setting timer
start_time = time.time()

#test = df.head(2)
```

```python
# Putting it all together 

#df["c_fname"] = ""
#df["c_lname"] = ""

def get_names(x):
    AF = x["AF"]
    #print AF
    #print type(AF)
    
    if (isinstance(AF, str)):
        try:
            authors_list = AF.split(";")
            corresponding = authors_list[-1]
    
            c_fname = corresponding.split(",")[1]
            c_lname = corresponding.split(",")[0]

            c_fname = c_fname.strip().lower()
            c_lname = c_lname.strip().lower()
    
            x["c_fname"] = c_fname
            x["c_lname"] = c_lname
        except:
            x["c_fname"] = "na"
            x["C_lname"] = "na"
    else:
        x["c_fname"] = "na"
        x["C_lname"] = "na"
    
    return x;

def get_gender(x):
    name = x["c_fname"]
    api_req = requests.get("https://api.genderize.io/?apikey=WOOOOOIT'SMYAPIKEYWOOOOOOOO=" + name)
    test = api_req.content
    if (test.find("null") == -1):
        response = eval(api_req.text)
        x["gender"] = response["gender"]
        x["probability"] = response["probability"]
        x["count"] = response["count"]
    else:
        x["gender"] = "na"
        x["probability"] = "na"
        x["count"] = "na"
    return x;

def get_citations(x):
    if ((x["gender"] != "na") and (x["probability"] >= 0.9)):
        query = scholarly.search_author(x["c_fname"] + " " + x["c_lname"])
        try:
            author = next(query).fill()
            try:
                x["scholar_id"] = author.id
            except:
                x["scholar_id"] = "na"
            try:
                x["citedby"] = author.citedby
            except:
                x["citedby"] = "na"
            #print x["citedby"]
        except StopIteration:
            x["scholar_id"] = "na"
            x["citedby"] = "na"
    else:
        x["scholar_id"] = "na"
        x["citedby"] = "na"
    return x;

tmp = []

def get_info(x):
    
    if (x["PD"] and x["PY"]):
        print x["PD"], x["PY"]
    
    get_names(x)
    get_gender(x)
    get_citations(x)
    
    tmp.append(x)
    
    return x;
 
#test = test.apply(get_info, axis=1)
```

```python
df = df.apply(get_info, axis=1)
df.to_csv("data/processed_recs4.csv")
```
