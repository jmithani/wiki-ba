```python
import pandas as pd

data = pd.read_csv("data/ratio_ptmn_len.csv", index_col="sample")
df = pd.DataFrame(data=data)
df
```

```python
def calculate_probability(x):
    x["prob_w"] = x["women"] / (1 - x["women"])
    x["prob_am"] = x["all_men"] / (1 - x["all_men"])
    x["prob_rm"] = x["random_av"] / (1 - x["random_av"])
    
    x["odds_rm"] = x["prob_rm"] / x["prob_w"]
    x["odds_am"] = x["prob_am"] / x["prob_w"]
    x["per_more_likely_rm"] = (x["odds_rm"] - 1) * 100
    x["per_more_likely_am"] = (x["odds_am"] - 1) * 100
    
    return x;

df = df.apply(calculate_probability, axis=0)
df
```
