```python
def get_wikipage(x):
    try:
        page = wk.page(x["c_fname"] + " " + x["c_lname"])
        x["page_id"] = page.pageid
        x["has_page"] = 1
        title = page.title
        x["page_title"] = title.encode('utf-8')

    except Exception as e: 
        print e
        
        x["page_id"] = "na"
        x["has_page"] = 0
        x["page_title"] = "na"
    
    try:
        nature = wk.page(x["c_fname"] + " " + x["c_lname"] + " Nature")
        x["page_id_nature"] = nature.pageid
        x["has_page_nature"] = 1
        title = page.title
        x["page_title_nature"] = title.encode('utf-8')
    except:
        x["page_id_nature"] = "na"
        x["has_page_nature"] = 0
        x["page_title_nature"] = "na"
        
    return x;

def match(x):
        
    pt = x["page_title"]
    ptn = x["page_title_nature"]
    lname = x["c_lname"]
    
    if (isinstance(pt, str) == True):
        pt = pt.lower()
        if (pt.find(lname) == -1):
            x["page_title_match"] = 0
        else:
            x["page_title_match"] = 1
    else:
        x["page_title_match"] = 0
    
    if (isinstance(ptn, str) == True):
        ptn = ptn.lower()
        if (ptn.find(lname) == -1):
            x["page_title_match_nature"] = 0
        else:
            x["page_title_match_nature"] = 1
    else:
        x["page_title_match"] = 0
    
    return x;

def get_wiki_info(x):
    get_wikipage(x)
    match(x)
    
    return x;
```

```python
women = women.apply(get_wiki_info, axis=1)
men = men.apply(get_wiki_info, axis=1)
```
